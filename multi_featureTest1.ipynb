# feature가 여러 개 일 때


```python
import tensorflow as tf
```

```python
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]
```

```python
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
y  = tf.placeholder(tf.float32)
```

```python
w1 = tf.Variable(tf.random_normal([1]), name='weight1')
w2 = tf.Variable(tf.random_normal([1]), name='weight2')
w3 = tf.Variable(tf.random_normal([1]), name='weight3')
b  = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y))
```

```python
train = tf.train.GradientDescentOptimizer(learning_rate=1e-5).minimize(cost)
```

```python
sess = tf.Session()
```

```python
# Variable들을 초기화
init = tf.global_variables_initializer()
```

```python
sess.run(init)
```

```python
for step in range(100000):
    _, cost_val, w1_val, w2_val, w3_val, b_val \
    = sess.run([train, cost, w1, w2, w3, b]  
    ,feed_dict={x1:x1_data, x2:x2_data, x3:x3_data, y:y_data }) 
    if step % 100 == 0:
        print(cost_val, w1_val, w2_val, w3_val, b_val)
```

```python
test1 = [80.] # test코드임. 예측을 정확히 하는 지 확인
test2 = [90.]
test3 = [70.]
```

```python
guess = sess.run(hypothesis, feed_dict={x1:test1, x2:test2, x3:test3})
```

```python
print(guess)
```

```python
sess.close()
```
