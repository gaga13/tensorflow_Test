```python
import tensorflow as tf
```

```python
x = tf.placeholder(tf.float32, shape=[None]) #shape=[개수, feature수]
y = tf.placeholder(tf.float32, shape=[None])
```

```python
W = tf.Variable(tf.random_normal([1]), name='weight') # 1 * 1의 행렬
# W의 쉐입은 x와 y가 결정한다
```

```python
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
hypothesis = x * W + b
```

```python
cost = tf.reduce_mean(tf.square(hypothesis-y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.01)
```

```python
train = optimizer.minimize(cost)
```

```python
sess = tf.Session()
```

```python
sess.run(tf.global_variables_initializer())
```

```python
for i in range(2000) :
    cost_val, W_val, b_val, _ = sess.run([cost, W, b, train], feed_dict={x:[1,2,3], y:[1.5, 2.5, 3.5]})
    if i % 20 == 0:
        print(cost_val, W_val, b_val)
        # 학습이 종료되면 cost_val이 최소값이 되어야 함
        # W : 1, b : 0.5 
    
```
