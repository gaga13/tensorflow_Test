# feature가 여러 개 일 때
- 2차원 Matrix로 처리하고자 함!

```python
import tensorflow as tf
```

```python
#x1_data = [73., 93., 89., 96., 73.]
#x2_data = [80., 88., 91., 98., 66.]
#x3_data = [75., 93., 90., 100., 70.]

#x = [[73, 80, 75], ...]

x_data = [[73., 93., 89., 96., 73.]
         ,[80., 88., 91., 98., 66.]
         ,[75., 93., 90., 100., 70.]]
y_data = [[152.], [185.], [180.]]
```

```python
#x1 = tf.placeholder(tf.float32)
#x2 = tf.placeholder(tf.float32)
#x3 = tf.placeholder(tf.float32)

x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
```

```python
#w1 = tf.Variable(tf.random_normal([1]), name='weight1')
#w2 = tf.Variable(tf.random_normal([1]), name='weight2')
#w3 = tf.Variable(tf.random_normal([1]), name='weight3')

w = tf.Variable(tf.random_normal([5, 1]), name='weight1')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
#hypothesis = x1 * w1 + x2 * w2 + x3 * w3 + b
hypothesis = tf.matmul(x, w) + b       #2차원으로 바꼈기 때문에 matmul사용
```

```python
cost = tf.reduce_mean(tf.square(hypothesis - y))
```

```python
train = tf.train.GradientDescentOptimizer(learning_rate=1e-5).minimize(cost)
```

```python
sess = tf.Session()
```

```python
# Variable들을 초기화
init = tf.global_variables_initializer()
```

```python
sess.run(init)
```

```python
for step in range(100000):
    _, cost_val, w_val, b_val = sess.run([train, cost, w, b],feed_dict={x:x_data, y:y_data}) 
    if step % 1000 == 0:
        print(cost_val, w_val, b_val)
```

```python
#test1 = [80.] # test코드임. 예측을 정확히 하는 지 확인
#test2 = [90.]
#test3 = [70.]

test = [[73., 93., 89., 96., 73.]]
```

```python
guess = sess.run(hypothesis, feed_dict={x: test})
```

```python
print(guess)
```
